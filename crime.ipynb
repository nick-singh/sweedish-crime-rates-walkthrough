```python
columns = df.columns
for c in [col for col in df.columns if 'theft' in col.lower:
    if c is not 'Year':
        if 'theft' in c.lower():
            plt.plot(df['Year'],df[c])
plt.xlabel('Year')
plt.ylabel('Number of thefts/robberies')
plt.legend()
```
